```python
import numpy as np
import pandas as pd
df=pd.read_csv('diabetes.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns=df.columns.str.strip().str.lower()
```

```python
# columns where 0 is not a valid medical value

invalid_zero_cols = ['glucose', 'bloodpressure', 'skinthickness', 'insulin', 'bmi']

#replace 0 with nan?
df[invalid_zero_cols] = df[invalid_zero_cols].replace(0, np.nan)
```

```python
df.isnull().sum()
```

```python
(df==0).any()
```

```python
# input missing values

for col in invalid_zero_cols:
  df[col].fillna(df[col].mean(), inplace=True)
```

```python
# median represent: a typical patient not too high not too low not influenced by outliers
```

```python
#feature target separation

#feature
x=df.drop(columns='outcome')
#target
y=df['outcome']
```

```python
x.columns
```

```python
y.name
```

```python
# train and test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
y.value_counts(normalize=True)
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

```python
# feature scaling - preporcessing

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
#  here using logistic regression classifire because it is a
#  binary classification problem, strong baseline for medical data

from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(x_train_scaled, y_train)

# here now the model trained, using scaled feature , can implement sadty
```

```python
# predictions
y_pred =log_reg.predict(x_test_scaled)
```

```python
y_pred[:10]
```

```python
log_reg.score(x_test_scaled, y_test)
# accuracy of test
```

```python
log_reg.score(x_train_scaled, y_train)
# accuracy of train
```

```python
y_prob = log_reg.predict_proba(x_test_scaled)
y_prob[:10]
```

```python
y_prob_diabetic=y_prob[:,1]
y_prob_diabetic[:10]
```

```python
# confusion matrix

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_pred)
cm

# [TN FP]
# [FN TP]

# TN - true negative - actully non diabetic
# FP - false positive - actully non diabetic but predicted diabetic
# FN - false negative - actully diabetic but predicted non diabetic
# TP - true positive - actully diabetic
```

```python
# classification report

from sklearn.metrics import classification_report
cr=classification_report(y_test, y_pred)
print(cr)

# here will see about precision , recall , f1-score , support
# precision : how correct the model was when it said 0 or 1
# recall: how many real 0 and 1 it found
# f1-score: balance score of precision and recall
# support: how many actual  0 and 1 patients

# exampl for weighted average:
# The weighted average in a classification report is an overall metric (for precision, recall, and f1-score)
# that takes into account the number of true instances for each class (the 'support').
```
